```python
import math
```

```python
def sigmoid(x):
  y = 1.0 / (1 + math.exp(-x))

def activate(inputs, weights):
  # perform net input
  h = 0 # stores the summation
  for x, w, in zip(inputs, weights):
    h += x*w

  # preform activation
  return sigmoid(h)
```

```python
if __name__ == "__main__":
  inputs = [.5, .3, .2]
  weights = [.4, .7, .2]
  output = activate(inputs, weights)
  print(output)
```
